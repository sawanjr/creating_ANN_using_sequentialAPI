```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation , Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
```

```python
train_label=[]
train_sample=[]
```

An experimental drug was tested on individuals from ages 13 to 100 in a clinic trail.

the trail had 2100 participants , half were under 65 years old, half were 65 year ild or older

around 95% of patient 65 or older experienced side effect

around 95% of patient under 65 experienced side effect

```python
# pip install scikit-learn
```

```python
from random import randint
for i in range(50):
    #5% of younger individuals who have experienced side effect
    random_younger=randint(13,64)
    train_sample.append(random_younger)
    train_label.append(1)
      
    #5% of older individuals who have not experienced side effect
    random_older=randint(65,100)
    train_sample.append(random_older)
    train_label.append(0)
    
for i in range(1000):
    #5% of younger individuals who have not experienced side effect
    random_younger=randint(13,64)
    train_sample.append(random_younger)
    train_label.append(0)
      
    #5% of older individuals who have experienced side effect
    random_older=randint(65,100)
    train_sample.append(random_older)
    train_label.append(1)
    
```

```python
import numpy as np
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler

train_label=np.array(train_label)
train_sample=np.array(train_sample)
train_label,train_sample=shuffle(train_label,train_sample)
```

```python
train_label
```

```python
train_sample
```

```python
scaler=MinMaxScaler(feature_range=(0,1))
# used to scale a set of data called train_sample. 
# Scaling is a technique used to transform numerical data to a standard range. 
# In this case, the data is being scaled so that all the values are 
# between 0 and 1.
# The MinMaxScaler class from the scikit-learn library is used for this purpose. 
# The feature_range parameter specifies the range to which 
# the data will be scaled, in this case, it is from 0 to 1.


scaled_train_samples=scaler.fit_transform(train_sample.reshape(-1,1))

# The train_sample data is reshaped to a single column using the 
# reshape() method. This is necessary because the scaler requires 
# the input data to be in a specific format.

# Finally, the fit_transform() method of the MinMaxScaler class is used to 
# apply the scaling transformation to the data. The resulting scaled data is 
# stored in the scaled_train_samples variable.



```

```python
scaled_train_samples # NOW VALUES HAVE BE STANDARDISE , BETWEEN 0 AND 1
```

```python

model = Sequential([
Dense(units=16,input_shape=(1,), activation='relu'),
    Dense(units=32, activation='relu'),
    Dense(units=2, activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate=0.0001), loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x=scaled_train_samples,y=train_label,batch_size=10,validation_split=0.1,epochs=30,verbose=2)
```

***as we can see our vaidation accuracy and training accuracy is nearly same***
It can be good if your validation accuracy and training accuracy are nearly the same, as it indicates that your model is not overfitting or underfitting.

Overfitting occurs when the model is too complex and learns the training data too well, resulting in high accuracy on the training data but poor generalization to new, unseen data. In contrast, underfitting occurs when the model is not complex enough to capture the underlying patterns in the data, resulting in low accuracy on both the training and validation data.

If your validation accuracy is significantly lower than your training accuracy, it may indicate overfitting, and you may need to consider reducing the complexity of your model, increasing the amount of regularization, or collecting more data.

On the other hand, if your training accuracy is significantly lower than your validation accuracy, it may indicate underfitting, and you may need to consider increasing the complexity of your model, improving your feature engineering, or collecting more data.
